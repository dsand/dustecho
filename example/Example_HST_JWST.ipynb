```python
#this is for 20 degrees

tobs = np.geomspace(6.20495675e+05,2.48198270e+09,num=100,endpoint=False)
#tobs
t_day = tobs/86400.
t_year = t_day/365.
#print(t_year)

#30 degrees
tobs_30deg = np.geomspace( 1.38102968e+06,5.52411872e+09,num=100,endpoint=False)
t_day_30deg = tobs_30deg/86400.
t_year_30deg = t_day_30deg/365.
```

```python
#equivalent of F110W
Lnu_f110 = np.loadtxt('nH1.0e+00_lam1.14um_theobs0.35_Ldnu_xcentr.txt',max_rows=1,skiprows=2)
fnu_f110_at10pc = Lnu_f110/(4.*np.pi*(10.*3.086E18)**2)

#equivalent of F140W
Lnu_f140 = np.loadtxt('nH1.0e+00_lam1.39um_theobs0.35_Ldnu_xcentr.txt',max_rows=1,skiprows=2)
fnu_f140_at10pc = Lnu_f140/(4.*np.pi*(10.*3.086E18)**2)

#equivalent of F160W
Lnu_f160 = np.loadtxt('nH1.0e+00_lam1.54um_theobs0.35_Ldnu_xcentr.txt',max_rows=1,skiprows=2)
fnu_f160_at10pc = Lnu_f160/(4.*np.pi*(10.*3.086E18)**2)

######JWST Nircam filters

#equivalent of F090W
Lnu_f090 = np.loadtxt('nH1.0e+00_lam0.90um_theobs0.35_Ldnu_xcentr.txt',max_rows=1,skiprows=2)
fnu_f090_at10pc = Lnu_f090/(4.*np.pi*(10.*3.086E18)**2)

#equivalent of F150W
Lnu_f150 = np.loadtxt('nH1.0e+00_lam1.50um_theobs0.35_Ldnu_xcentr.txt',max_rows=1,skiprows=2)
fnu_f150_at10pc = Lnu_f150/(4.*np.pi*(10.*3.086E18)**2)

#equivalent of F090W; 30 deg
Lnu_30deg_f090 = np.loadtxt('nH1.0e+00_lam0.90um_theobs0.52_Ldnu_xcentr.txt',max_rows=1,skiprows=2)
fnu_30deg_f090_at10pc = Lnu_30deg_f090/(4.*np.pi*(10.*3.086E18)**2)

#equivalent of F150W; 30 deg
Lnu_30deg_f150 = np.loadtxt('nH1.0e+00_lam1.50um_theobs0.52_Ldnu_xcentr.txt',max_rows=1,skiprows=2)
fnu_30deg_f150_at10pc = Lnu_30deg_f150/(4.*np.pi*(10.*3.086E18)**2)


#equivalent of F090W; low nH
Lnu_lonH_f090 = np.loadtxt('nH1.0e-01_lam0.90um_theobs0.35_Ldnu_xcentr.txt',max_rows=1,skiprows=2)
fnu_lonH_f090_at10pc = Lnu_lonH_f090/(4.*np.pi*(10.*3.086E18)**2)

#equivalent of F150W; low nH
Lnu_lonH_f150 = np.loadtxt('nH1.0e-01_lam1.50um_theobs0.35_Ldnu_xcentr.txt',max_rows=1,skiprows=2)
fnu_lonH_f150_at10pc = Lnu_lonH_f150/(4.*np.pi*(10.*3.086E18)**2)


#equivalent of F200W
Lnu_f200 = np.loadtxt('nH1.0e+00_lam2.00um_theobs0.35_Ldnu_xcentr.txt',max_rows=1,skiprows=2)
fnu_f200_at10pc = Lnu_f200/(4.*np.pi*(10.*3.086E18)**2)

#equivalent of F277W
Lnu_f277 = np.loadtxt('nH1.0e+00_lam2.77um_theobs0.35_Ldnu_xcentr.txt',max_rows=1,skiprows=2)
fnu_f277_at10pc = Lnu_f277/(4.*np.pi*(10.*3.086E18)**2)

#equivalent of F277W; 30 deg
Lnu_30deg_f277 = np.loadtxt('nH1.0e+00_lam2.77um_theobs0.52_Ldnu_xcentr.txt',max_rows=1,skiprows=2)
fnu_30deg_f277_at10pc = Lnu_30deg_f277/(4.*np.pi*(10.*3.086E18)**2)

#equivalent of F277W; lo nH
Lnu_lonH_f277 = np.loadtxt('nH1.0e-01_lam2.77um_theobs0.35_Ldnu_xcentr.txt',max_rows=1,skiprows=2)
fnu_lonH_f277_at10pc = Lnu_lonH_f277/(4.*np.pi*(10.*3.086E18)**2)

#equivalent of F356W
Lnu_f356 = np.loadtxt('nH1.0e+00_lam3.56um_theobs0.35_Ldnu_xcentr.txt',max_rows=1,skiprows=2)
fnu_f356_at10pc = Lnu_f356/(4.*np.pi*(10.*3.086E18)**2)

#equivalent of F444W
Lnu_f444 = np.loadtxt('nH1.0e+00_lam4.44um_theobs0.35_Ldnu_xcentr.txt',max_rows=1,skiprows=2)
fnu_f444_at10pc = Lnu_f444/(4.*np.pi*(10.*3.086E18)**2)


```

```python
DM=33.05 #from Cantiello+18
magabs_f150 = -2.5*np.log10(fnu_f150_at10pc)-48.60
mag_f150 = magabs_f150+DM

magabs_f140 = -2.5*np.log10(fnu_f140_at10pc)-48.60
mag_f140 = magabs_f140+DM

magabs_f160 = -2.5*np.log10(fnu_f160_at10pc)-48.60
mag_f160 = magabs_f160+DM

######JWST Nircam filters

magabs_f090 = -2.5*np.log10(fnu_f090_at10pc)-48.60
mag_f090 = magabs_f090+DM

magabs_lonH_f090 = -2.5*np.log10(fnu_lonH_f090_at10pc)-48.60
mag_lonH_f090 = magabs_lonH_f090+DM

magabs_30deg_f090 = -2.5*np.log10(fnu_30deg_f090_at10pc)-48.60
mag_30deg_f090 = magabs_30deg_f090+DM

magabs_f150 = -2.5*np.log10(fnu_f150_at10pc)-48.60
mag_f150 = magabs_f150+DM

magabs_30deg_f150 = -2.5*np.log10(fnu_30deg_f150_at10pc)-48.60
mag_30deg_f150 = magabs_30deg_f150+DM

magabs_lonH_f150 = -2.5*np.log10(fnu_lonH_f150_at10pc)-48.60
mag_lonH_f150 = magabs_lonH_f150+DM

magabs_f200 = -2.5*np.log10(fnu_f200_at10pc)-48.60
mag_f200 = magabs_f200+DM

magabs_f277 = -2.5*np.log10(fnu_f277_at10pc)-48.60
mag_f277 = magabs_f277+DM

magabs_30deg_f277 = -2.5*np.log10(fnu_30deg_f277_at10pc)-48.60
mag_30deg_f277 = magabs_30deg_f277+DM

magabs_lonH_f277 = -2.5*np.log10(fnu_lonH_f277_at10pc)-48.60
mag_lonH_f277 = magabs_lonH_f277+DM

magabs_f356 = -2.5*np.log10(fnu_f356_at10pc)-48.60
mag_f356 = magabs_f356+DM

magabs_f444 = -2.5*np.log10(fnu_f444_at10pc)-48.60
mag_f444 = magabs_f444+DM

```

```python
fig = plt.figure(figsize=(12,12))
plt.rcParams.update({'font.size':24})

plt.xlabel(r'${\rm Phase}$ $[yr]$',fontsize=26)
plt.ylabel(r'$\rm Apparent\ Magnitude\ (AB mag)$',fontsize=26)
plt.plot(t_year,mag_f110,'-',ms=8,alpha=0.6,mew=0.5,color='blue', label='F110W')
plt.plot(t_year,mag_f140,'-',ms=8,alpha=0.6,mew=0.5,color='green', label='F140W')
plt.plot(t_year,mag_f160,'-',ms=8,alpha=0.6,mew=0.5,color='red', label='F160W')
plt.plot(t_year,mag_f200,'-',ms=8,alpha=0.6,mew=0.5,color='purple', label='F200W')
plt.xlim([0.,8.])
plt.ylim([30.0,25.0])
#plt.xlim(0.,10.)
#plt.ylim(30.0,25.0)
plt.legend(loc='upper right',markerscale=1,numpoints=1,ncol=3,labelspacing=0.15,handlelength=1.0,fontsize=20)
```
